```python
high_corr_cols = ['emp.var.rate','euribor3m','nr.employed']
```

```python
bank1 = bank.copy()
bank1.columns
```

```python
bank1.drop(high_corr_cols,inplace=True,axis=1)  # axis=1 indicates columns
bank1.columns
```

```python
bank1.shape
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
bank_encoded = bank1.apply(lb.fit_transform)
bank_encoded
```

```python
bank_encoded['y'].value_counts()
```

```python
x = bank_encoded.drop('y',axis=1)  # independent variable
y = bank_encoded['y']              # dependent variable
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

```python
from sklearn.model_selection import train_test_split

print(4119*0.25)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

def eval_model(y_test,y_pred):
    acc = accuracy_score(y_test,y_pred)
    print('Accuracy_Score',acc)
    cm = confusion_matrix(y_test,y_pred)
    print('Confusion Matrix\n',cm)
    print('Classification Report\n',classification_report(y_test,y_pred))

def mscore(model):
    train_score = model.score(x_train,y_train)
    test_score = model.score(x_test,y_test)
    print('Training Score',train_score)  
    print('Testing Score',test_score)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(criterion='gini',max_depth=5,min_samples_split=10)
dt.fit(x_train,y_train)
```

```python
mscore(dt)
```

```python
ypred_dt = dt.predict(x_test)
print(ypred_dt)
```

```python
eval_model(y_test,ypred_dt)
```

```python
from sklearn.tree import plot_tree
cn = ['no','yes']
fn = x_train.columns
print(fn)
print(cn)
```

```python
plot_tree(dt,class_names=cn,filled=True)
plt.show()
```

```python
dt1 = DecisionTreeClassifier(criterion='entropy',max_depth=4,min_samples_split=15)
dt1.fit(x_train,y_train)
```

```python
mscore(dt1)
```

```python
ypred_dt1 = dt1.predict(x_test)
```

```python
eval_model(y_test,ypred_dt1)
```
